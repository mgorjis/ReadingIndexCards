```python
# Create a list of participant in this dictionary form: (graph name: unique id )


ListOfParticipant=[]
ListOfLocation=[]
#ListOfUniqueid=[]
ElementType=[]
ListOfInfluence_positive=[]
ListOfInfluence_negative=[]
ParticipantsDict={}
ListOfEvents=[]

def isNaN(num):
    return num != num


import math
import re
nan=float("nan")

for i in range(2,172):
    
    name=df['Graph model variable name'][i].replace(' ','')
    name_short=name
    if name_short.endswith('nuc'): name_short=name_short.replace('nuc',''); 
    if name_short.endswith('cyto'): name_short=name_short.replace('cyto','')
    if name_short.endswith('Ract'): name_short=name_short.replace('Ract','')
    if name_short.endswith('rna'): name_short=name_short.replace('rna','')
    if name_short.endswith('gene'): name_short=name_short.replace('gene','')
    if name_short.endswith('pm'): name_short=name_short.replace('pm','')
    if name_short.endswith('ex'): name_short=name_short.replace('ex','')
    if name_short.endswith('act'): name_short=name_short.replace('act','')
    ListOfParticipant.append(name)
    
    location=df['Location'][i]
        
    ID=df['Unique ID'][i]
    if (':' in ID) == False:
        ID='uniprot:'+ID
    ParticipantsDict.update({name_short:ID})
    
    IP=df['Influence set'][i]  #Influence_positive
    if type(IP)==str:
        
        IP=IP.replace('(MAC)',' ')
 
        IP=IP.replace(';',' ').replace(',',' ').replace(' or ',' ').replace(' and ',' ').replace('{',' ').replace('}',' ').replace('[',' ').replace(']',' ').replace('(',' ').replace(')',' ').split()
        
        for j in IP:
            IP_short=j
            
            if IP_short.endswith('nuc'): IP_short=IP_short.replace('nuc',''); 
            if IP_short.endswith('cyto'): IP_short=IP_short.replace('cyto','')
            if IP_short.endswith('Ract'): IP_short=IP_short.replace('Ract','')
            if IP_short.endswith('rna'): IP_short=IP_short.replace('rna','')
            if IP_short.endswith('gene'): IP_short=IP_short.replace('gene','')
            if IP_short.endswith('pm'): IP_short=IP_short.replace('pm','')
            if IP_short.endswith('ex'): IP_short=IP_short.replace('ex','')
            if IP_short.endswith('act'): IP_short=IP_short.replace('act','')
            ListOfInfluence_positive.append(IP_short)
        
        
    IN=df['Unnamed: 28'][i]    #Influence_negative
    
    if type(IN)==str:
        
        IN=IN.replace('(MAC)',' ')
     
        IN=IN.replace(';',' ').replace(',',' ').replace(' or ',' ').replace(' and ',' ').replace('{',' ').replace('}',' ').replace('[',' ').replace(']',' ').replace('(',' ').replace(')',' ').split()
        
        for j in IN:
            IN_short=j
            
            if IN_short.endswith('nuc'): IN_short=IN_short.replace('nuc','');
            if IN_short.endswith('cyto'): IN_short=IN_short.replace('cyto','')
            if IN_short.endswith('Ract'): IN_short=IN_short.replace('Ract','')
            if IN_short.endswith('rna'): IN_short=IN_short.replace('rna','')
            if IN_short.endswith('gene'): IN_short=IN_short.replace('gene','')
            if IN_short.endswith('pm'): IN_short=IN_short.replace('pm','')
            if IN_short.endswith('ex'): IN_short=IN_short.replace('ex','')
            if IN_short.endswith('act'): IN_short=IN_short.replace('act','')
            ListOfInfluence_negative.append(IN_short)
        
    ET=df['Unnamed: 26'][i]  #Event Type
    ET=str(ET).replace(',',' ').split()
    
    
 
    k=0
    if not isNaN(IP):
        for j in range(0,len(IP)):
            if ET[0]=='nan': 
                ET1='?' #nan' 
            else:
                ET1=ET[k]
                
            ListOfEvents.append((IP[j],name,'Positive',ET1,location))
            k=k+1
    if not isNaN(IN):
        for j in range(0,len(IN)):
            if ET[0]=='nan': 
                ET1='?' #'nan' 
            else:
                ET1=ET[k]
                
            ListOfEvents.append((IN[j],name,'Negative',ET1,location))
            k=k+1
```

### Also, we are creating a dictionary that map each name to its unique Id. We have read the columns ('Graph model variable name' and 'Unique ID' ) from excel file. Also, for making the dictionary complete, we read also from NMZ-CMU_160111.tsv

```python
import csv
path="/Users/mgorjise/Desktop/new/CMU/NMZ-CMU_160111.tsv"
#path="/Users/mgorjise/Documents/new/reading/CMU/NMZ-CMU_160111.tsv"

with open(path, encoding='ISO-8859-1') as tsvin:
    next(tsvin, None)
    next(tsvin, None)
    next(tsvin, None)
    for line in csv.reader(tsvin, dialect="excel", delimiter='\t',skipinitialspace=True):
        ParticipantsDict.update({line[0]:line[1]+':'+line[2]})
ParticipantsDict.update({"ROS":"chemidplus:D017382000"})
```
