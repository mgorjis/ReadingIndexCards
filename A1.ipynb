Look in all the folders started with "P", and file all the json files in them, Then generate a big list of json files' paths

```python
import os
import glob

BigList=[]
for infile in glob.glob( os.path.join(path, 'P*') ):
    newlist=glob.glob(os.path.join(infile, '*.json'))
    BigList=BigList+newlist

```

```python
#Number of processed json files
NumberOfFiles=len(BigList)    
print(NumberOfFiles)
```

```python
import json

ListOfCellularComponent=[]
ListOfOrgans=[]
ListOfSpecies=[]
ListOfParticipantA=[]
ListOfParticipantB=[]
ListOfEvents=[]
ListOfAddMoification=[]
ListOfInhibitMoification=[]

length=NumberOfFiles
for filename in BigList[0:NumberOfFiles]:  #length
    json_data=open(filename).read()
    data = json.loads(json_data)
    
    a=data.get('extracted_information').get('context')
    if a !=None:
        b=a.get('Organ')
        if  b !=None:
            ListOfOrgans=ListOfOrgans + b
    
    a=data.get('extracted_information').get('context')
    if a !=None:
        b=a.get('Cellular_component')
        if  b !=None:
            ListOfCellularComponent=ListOfCellularComponent + b
            
    a=data.get('extracted_information').get('context')
    if a !=None:
        b=a.get('Species')
        if  b !=None:
            ListOfSpecies=ListOfSpecies + b
            
    a=data.get('extracted_information').get('participant_a')
    if a != None: 
        if type(a)==list:
            NumberOfParticipantA=len(a)
            for i in range(1,NumberOfParticipantA) :
                b=a[i].get('identifier')
                ListOfParticipantA.append(b)
        if  type(a)==dict:
            b=a.get('identifier')
            ListOfParticipantA.append(b)
  
            
    a=data.get('extracted_information').get('participant_b')
    b=a.get('identifier')
    ListOfParticipantB.append(b)       
            
    a=data.get('extracted_information').get('interaction_type')
    if a=="adds_modification":
        ListOfAddMoification.append(data.get('extracted_information').get('modifications')[0].get('modification_type'))
        
    if a=="inhibits_modification":
        ListOfInhibitMoification.append(data.get('extracted_information').get('modifications')[0].get('modification_type'))
        
    ListOfEvents.append(a)
```
